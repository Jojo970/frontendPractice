```python
import numpy as np
import pandas as pd
```

```python
mp_df = pd.read_csv('file_1.csv', encoding = "ISO-8859-1")
recno_df = pd.read_csv('file_2.csv', encoding = "ISO-8859-1")
```

```python
mp_df.columns
```

```python
mp_df.sort_values(by = ['RecNo '], inplace = True)

recno_df.sort_values(by = ['recno'], inplace = True)
```

```python
n=1

for x, y in recno_df['recno'].iteritems():
    new_df = mp_df[mp_df['RecNo '].isin([y])]
    break
    
for x, y in recno_df['recno'].iteritems():
    check = mp_df['RecNo '].isin([y])
    if check.any() == True:
        to_append = mp_df[mp_df['RecNo '].isin([y])]
        new_df = new_df.append(to_append)
        
    else:
        n+=1
        new_df = new_df.append(pd.Series(), ignore_index = True)
        new_df.fillna(value = n, inplace =True)


new_df.columns

real_df = new_df[['RecNo ', 'Adj Amt', 'Carrier ID']]

print(real_df.head(100))
```
