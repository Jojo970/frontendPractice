```python
import numpy as np

import pandas as pd
```

```python
df_load_id = pd.read_csv('load_id_list.csv', encoding = 'unicode_escape')

df_recno = pd.read_csv('pro_numbers.csv')

df_container = pd.read_csv('pro_numbers_2.csv')
```

```python
df_load_id.drop_duplicates(inplace = True)

df_load_id.head(5)
```

```python
df_recno.head(5)
```

```python
df_container.head()
```

```python
n = 0

for x, y in df_container['container'].iteritems():
    new_df = df_load_id[df_load_id['Container#'].isin([y])]
    break
    
for x, y in df_container['container'].iteritems():
    check = df_load_id['Container#'].isin([y])
    if check.any() == True:
        to_append = df_load_id[df_load_id['Container#'].isin([y])]
        new_df = new_df.append(to_append)
        
        
    else:
        n+=1
        new_df = new_df.append(pd.Series(), ignore_index = True)
        new_df.fillna(value = n, inplace =True)


new_df.columns

real_df = new_df[['Load ID']]

real_df.head(10)
```

```python
with pd.ExcelWriter('yst_out.xlsx') as writer:
    real_df.to_excel(writer, sheet_name='Sheet 1')
    
```
