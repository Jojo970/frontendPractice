```python
import numpy as np
import pandas as pd

finder_df = pd.read_csv('containers.csv')

look_in_df = pd.read_csv('look_in.csv')
```

```python
look_in_df.head(5)
```

```python
for x, y in finder_df['Container#'].iteritems():
    new_df = look_in_df[look_in_df['Container No'].isin([y])]
    break

for x , y in finder_df['Container#'].iteritems():
    check = look_in_df['Container No'].isin([y])
    if check.any() == True:
        to_append = look_in_df[look_in_df['Container No'].isin([y])]
        new_df = new_df.append(to_append)

new_df.head(5)
```

with pd.ExcelWriter('pros_out.xlsx') as writer:
    new_df.to_excel(writer, sheet_name='Sheet 1')
